```python
# Install required dependencies
!pip install -U torch torchvision opencv-python-headless
!pip install 'git+https://github.com/facebookresearch/detectron2.git'

import os
import cv2
import random
import numpy as np
import torch
from google.colab.patches import cv2_imshow
from detectron2 import model_zoo
from detectron2.engine import DefaultTrainer, DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer, ColorMode
from detectron2.data import MetadataCatalog, DatasetCatalog

# Download COCO dataset into /content/sample_data
!mkdir -p /content/sample_data/coco
!wget http://images.cocodataset.org/zips/val2017.zip -P /content/sample_data/coco/
!unzip -q /content/sample_data/coco/val2017.zip -d /content/sample_data/coco/
!wget http://images.cocodataset.org/annotations/annotations_trainval2017.zip -P /content/sample_data/coco/
!unzip -q /content/sample_data/coco/annotations_trainval2017.zip -d /content/sample_data/coco/

# Register COCO Dataset
from detectron2.data.datasets import register_coco_instances

coco_image_path = "/content/sample_data/coco/val2017"
coco_annotation_path = "/content/sample_data/coco/annotations/instances_val2017.json"

# Register the COCO dataset with Detectron2
register_coco_instances("coco_val_instance", {}, coco_annotation_path, coco_image_path)
coco_metadata = MetadataCatalog.get("coco_val_instance")
dataset_dicts = DatasetCatalog.get("coco_val_instance")

# Configure the Mask R-CNN model for instance segmentation
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.DATASETS.TRAIN = ("coco_val_instance",)  # Since we are using the validation set for demonstration
cfg.DATASETS.TEST = ()
cfg.DATALOADER.NUM_WORKERS = 2
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 500  # Number of iterations for demonstration
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 80  # 80 classes in COCO dataset

# Train the model (for demonstration purposes, training is done for a few iterations)
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()

# Load trained model for inference
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # Set the threshold for predictions
cfg.DATASETS.TEST = ("coco_val_instance",)
predictor = DefaultPredictor(cfg)

# Make predictions and visualize the results
for d in random.sample(dataset_dicts, 3):  # Randomly select 3 images to visualize
    img_path = d["file_name"]
    img = cv2.imread(img_path)
    outputs = predictor(img)

    # Visualize the predicted masks
    v = Visualizer(img[:, :, ::-1], metadata=coco_metadata, scale=0.8, instance_mode=ColorMode.IMAGE_BW)
    out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
    cv2_imshow(out.get_image()[:, :, ::-1])
```
